```python
#encodes 'Yes'=1,'No'=0 for Attrition column
from sklearn.preprocessing import LabelEncoder
encoder1 = LabelEncoder()
dataset['Attrition'] = encoder1.fit_transform(dataset['Attrition'])

#encodes 'Travel_Rarely'=2,'Travel_Frequently'=1,'Non_Travel'=0 for BusinessTravel column
from sklearn.preprocessing import LabelEncoder
encoder2 = LabelEncoder()
dataset['BusinessTravel'] = encoder2.fit_transform(dataset['BusinessTravel'])

#encodes 'Sales'=2,'Research & Development'=1,'Human Resources'=0 for Department column
from sklearn.preprocessing import LabelEncoder
encoder3 = LabelEncoder()
dataset['Department'] = encoder3.fit_transform(dataset['Department'])

#encodes 'Technical Degree'=5,'Other'=4,'Medical'=3,'Marketing'=2,'Life Sciences'=1,
#'Human Resources'=0 for EducationField column
from sklearn.preprocessing import LabelEncoder
encoder4 = LabelEncoder()
dataset['EducationField'] = encoder4.fit_transform(dataset['EducationField'])

#encodes 'Male'=1,'Female'=0 for Gender column
from sklearn.preprocessing import LabelEncoder
encoder5 = LabelEncoder()
dataset['Gender'] = encoder5.fit_transform(dataset['Gender'])

#encodes 'Sales Representative'=8,'Sales Executive'=7,'Research Scientist'=6,'Research Director'=5,
#'Manufacturing Director'=4,'Manager'=3,'Laboratory Technician'=2,'Human Resources'=1,
#'Healthcare Representative'=0 in JobRole column
from sklearn.preprocessing import LabelEncoder
encoder6 = LabelEncoder()
dataset['JobRole'] = encoder6.fit_transform(dataset['JobRole'])

#encodes 'Single'=2,'Married'=1,'Divorced'=0 in MaritalStatus column
from sklearn.preprocessing import LabelEncoder
encoder7 = LabelEncoder()
dataset['MaritalStatus'] = encoder7.fit_transform(dataset['MaritalStatus'])

from sklearn.preprocessing import LabelEncoder
encoder8 = LabelEncoder()
dataset['Over18'] = encoder8.fit_transform(dataset['Over18'])
```

# correlation b/w attrition and age

```python
stats,p=pearsonr(dataset.Attrition,dataset.Age)
print("Correlation:",stats,"\n P-value:",p)
if p>=0.05:
    print("\nAccept Null Hypothesis")
else:
    print("\nReject Null Hypothesis")
```
