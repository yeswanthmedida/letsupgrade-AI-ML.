```python
# Encoding Our objective was to find out the factors which need to be control to curb the attrition of employee's, as the Attrition is a categorical variable it's needed to be encoded so as to proceed with further analysis,here two such encoding are employed,namely- 1) Label encoding 2) One-Hot encodingLabel encoding Label encoding simply assign numeric values to different categories starting from 0, here we are encoding only Attrition field where 'Yes'=1 and 'No'=0.
```

```python
from sklearn.preprocessing import LabelEncoder
encoder1 = LabelEncoder()
dataset2['Attrition'] = encoder1.fit_transform(dataset2['Attrition'])
```

# Label encoding of all the other categorical columns

```python
from sklearn.preprocessing import LabelEncoder
encoder2 = LabelEncoder()
dataset3['BusinessTravel'] = encoder2.fit_transform(dataset3['BusinessTravel'])

#encodes 'Sales'=2,'Research & Development'=1,'Human Resources'=0 for Department column
from sklearn.preprocessing import LabelEncoder
encoder3 = LabelEncoder()
dataset3['Department'] = encoder3.fit_transform(dataset3['Department'])

#encodes 'Technical Degree'=5,'Other'=4,'Medical'=3,'Marketing'=2,'Life Sciences'=1,
#'Human Resources'=0 for EducationField column
from sklearn.preprocessing import LabelEncoder
encoder4 = LabelEncoder()
dataset3['EducationField'] = encoder4.fit_transform(dataset3['EducationField'])

#encodes 'Male'=1,'Female'=0 for Gender column
from sklearn.preprocessing import LabelEncoder
encoder5 = LabelEncoder()
dataset3['Gender'] = encoder5.fit_transform(dataset3['Gender'])

#encodes 'Sales Representative'=8,'Sales Executive'=7,'Research Scientist'=6,
#'Research Director'=5,'Manufacturing Director'=4,'Manager'=3,'Laboratory Technician'=2,'Human Resources'=1,
#'Healthcare Representative'=0 in JobRole column
from sklearn.preprocessing import LabelEncoder
encoder6 = LabelEncoder()
dataset3['JobRole'] = encoder6.fit_transform(dataset3['JobRole'])

#encodes 'Single'=2,'Married'=1,'Divorced'=0 in MaritalStatus column
from sklearn.preprocessing import LabelEncoder
encoder7 = LabelEncoder()
dataset3['MaritalStatus'] = encoder7.fit_transform(dataset3['MaritalStatus'])
```
