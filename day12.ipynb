# Problem Statement 2
## H0: There is no significant differnce in distance from home between Attrition(Yes) and Attrition(No).(Null hypothesis)

H1: There is significant differnce in distance from home between Attrition(Yes) and Attrition(No).(Alternate hypothesis)

Attrition can take up value as either "Yes" or "No", this give us two samples-

1st Sample- Employee with Attrition as "Yes" with certain distance from home values.

2nd Sample- Employee with Attrition as "No" with certain distance from home values.

Both these samples are independent so we have to apply Mann Whitney T


```python
from scipy.stats import mannwhitneyu
stats2,p2=mannwhitneyu(df2.DistanceFromHome,df3.DistanceFromHome)
print("Critical value:",stats2)
print("p value:",p2)
```
